```python
import pandas as pd
%matplotlib inline
```

Loading in the data from csv file, removing rows with no  neighborhood data, displaying them in descending order

```python
data = pd.read_csv("Schools.csv")
Schools_df = pd.DataFrame(data)

Schools_df = Schools_df[Schools_df['location_city'].str.len() > 0]
Schools_df = Schools_df.sort_values(by=["location_city"], ascending=True)

Schools_counts = Schools_df['location_city'].value_counts()
print(Schools_counts.head(10))

Schools_counts.plot(kind='bar')
```

This does work but we still need to find the best private schools in a neighborhood and not only the city. i am going to make it so i can figure out which neigborhood gives the best education.

```python
data = pd.read_csv("Schools.csv")
Schools_df = pd.DataFrame(data)

Schools_df = Schools_df[Schools_df['category'].str.len() > 0]
Schools_df = Schools_df.sort_values(by=["category"], ascending=True)

Schools_counts = Schools_df['category'].value_counts()
print(Schools_counts.head(10))

Schools_counts.plot(kind='bar')
```
