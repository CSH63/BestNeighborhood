```python
import pandas as pd
df = pd.read_csv("SmartTrash.csv")
df.groupby(['neighborhood'])
Neighborhoods = df['neighborhood'].value_counts()
Neighborhoods = Neighborhoods.head(10)
graph = Neighborhoods.plot.barh(x="Neighborhood",y="Number of Cans",figsize=(15,Neighborhoods.shape[0] * 0.3), legend=None)
graph.set_xlabel("Number of smart cans in the area")
graph.get_figure().savefig('SmartCans.pdf')
```
